```python
import requests
import pandas as pd
import time
from tqdm import tqdm
```

```python
def web_request(method_name, url, is_urlencoded=True, timeout_seconds=3):
    """Web GET or POST request를 호출 후 그 결과를 dict형으로 반환 """
    method_name = method_name.upper() # 메소드이름을 대문자로 바꾼다 
    if method_name not in ('GET', 'POST'):
        raise Exception('method_name is GET or POST plz...')
        
    if method_name == 'GET': # GET방식인 경우
        response = requests.get(url=url, timeout=timeout_seconds)
    elif method_name == 'POST': # POST방식인 경우
        if is_urlencoded is True:
            response = requests.post(url=url, \
                                     timeout=timeout_seconds, headers=header)
        else:
            response = requests.post(url=url, \
                                     timeout=timeout_seconds, headers=header)
    return response
    
def web_request_retry(num_retry=3, sleep_seconds=1, **kwargs):
    """timeout발생 시 sleep_seconds쉬고 num_retyrp번 재시도 한다"""
    for n in range(num_retry):
        try:
            return web_request(**kwargs)
        except:
          time.sleep(sleep_seconds)
          continue
        # except requests.exceptions.Timeout:
        #     print(str(n+1) + ' Timeout')
        #     time.sleep(sleep_seconds)
        #     continue
        # except requests.exceptions.ConnectionError:
        #     time.sleep(sleep_seconds)
        #     print("Connection Error!!!!!!!!")
        #     continue
    return None

def checkCodeName(code, code_name):
  url = f'https://api.deepsearch.com/v1/compute?input=FindEntityByName("Financial","{code_name}")'
  res = web_request_retry(method_name='POST', url=url, is_urlencoded=True, num_retry=5 )
  res_data = res.json()

  try:
    if code in res_data['data']['pods'][1]['content']['data']['symbol']:
      return True
    else:
      return False
  except:
    return False
```

```python
for stock in stock_total_list:
  stock_code = "KRX:" + stock[0].zfill(6)
  stock_name = stock[1]

  keyword_list = []

  if stock_name in stock_list:
    # if stock_name == '롯데제과':
      keyword_original = df[stock_name].values.tolist()
      for keyword in tqdm(keyword_original):
        if not pd.isna(keyword):
          if '&' in keyword:
            search_keyword = keyword.replace('&', '%26')
          else:
            search_keyword = keyword
            
          if '#' in keyword:
            search_keyword = search_keyword.replace('#', '%23')

          if '"' in keyword:
            search_keyword = search_keyword.replace('"', "'")

          
          url = f'https://api.deepsearch.com/v1/compute?input=DocumentAggregation("news","economy","{search_keyword}","named_entities.entities.company.name:10",date_from=2010-01-01)'
          res = web_request_retry(method_name='POST', url=url, is_urlencoded=True, num_retry=5 )
          res_data = res.json()

          try:
            len_data = len(res_data['data']['pods'][1]['content']['data']['key'])

            if len_data != 0:
              related_company = res_data['data']['pods'][1]['content']['data']['key'][0]
              if '&' in related_company:
                related_company = related_company.replace('&', '%26')

              if checkCodeName(code = stock_code, code_name = related_company):
                keyword_list.append(keyword)
                print("keyword: " + keyword + " !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")

          except Exception as e:
            print("error : ", e)
            print("keyword : " , keyword)
          
      keyword_list = pd.Series(keyword_list)

      result = pd.DataFrame({
          '키워드' : keyword_list
      })

      result.to_csv(f'/content/drive/MyDrive/KAKAOPAYSEC/data_total/{stock_name}_13ver.csv', index=False, encoding='utf-8-sig')
```
